```python
#Which country do you currently work in?
import pandas as pd
from ipynb.fs.full.utils import get_df, rename_values
from ipynb.fs.full.plots import single_barplot
```

```python
df = get_df()
df = df[['D2_Country']]
df.head(10)
```

```python


# index of our barplot
indexes = list(df['D2_Country'].value_counts().index)
# total answers of each index
values = list(df['D2_Country'].value_counts())
# dataframe specific for question D3
d2_df= pd.DataFrame(data={'Paises': indexes, 'Answers': values})
```

```python
single_barplot(d2_df,'Paises', 'Answers', 
               title="Which country do you currently work in?",
               x_label="Paises", y_label="Answers", x_label_rotation=60)

```

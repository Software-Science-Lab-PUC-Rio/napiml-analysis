```python
# get formatted and ready to use dataframe
df = get_df()
```

```python
# q7 dataframe not considering the answers with no value
q7_df = pd.DataFrame()
q7_df['ML Solution Optimality'] = list(df['Q7_ML_Solution_Optimality_Extra_Effort'])
```

```python
q7_df['ML Solution Optimality'].unique()
```

```python
q7_df['ML Solution Optimality'].value_counts()
```

```python
# remove those who didn't answered
q7_df = q7_df[q7_df['ML Solution Optimality'] != '-77']
```

```python
# convert column to integer - if you read before this line, you will get an string like '14'
q7_df = q7_df.astype({"ML Solution Optimality": int}, errors='raise') 
```

```python
single_boxplot(q7_df, "ML Solution Optimality",
               title="To what degree do you think that these “suboptimal” solutions cause added effort in the evolution of the ML-enabled system over time on a scale from 0% means no added effort/time needed, to 100% meaning a doubling of effort/time.")
```

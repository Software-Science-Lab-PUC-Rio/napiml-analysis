```python
import pandas as pd
from ipynb.fs.full.utils import get_df
from ipynb.fs.full.plots import single_boxplot, single_histplot, single_violinplot
```

```python
# get formatted and ready to use dataframe
df = get_df()
```

```python
df.head(10)
```

```python
# d6 dataframe with only one column
d6_df = pd.DataFrame()
d6_df['ML Experience'] = list(df['D6_ML_Experience'])
```

```python
# remove those who didn't answered
d6_df = d6_df[d6_df['ML Experience'] != '-99']
```

```python
# convert column to integer - if you read before this line, you will get an string like '14'
d6_df = d6_df.astype({"ML Experience": int}, errors='raise') 
```

```python
single_boxplot(d6_df, "ML Experience",
               title="How many years of experience do you have in developing ML-enabled systems?")
```

```python
single_boxplot(d6_df, "ML Experience", outliers=False,
               title="How many years of experience do you have in developing ML-enabled systems?")
```

```python
single_violinplot(d6_df, "ML Experience",
                  title="How many years of experience do you have in developing ML-enabled systems?")
```

```python
single_histplot(d6_df, "ML Experience",
                title="How many years of experience do you have in developing ML-enabled systems?")
```

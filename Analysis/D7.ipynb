```python
import pandas as pd
from ipynb.fs.full.utils import get_df
from ipynb.fs.full.plots import single_boxplot, single_histplot, single_violinplot
```

```python
# get formatted and ready to use dataframe
df = get_df()
```

```python
df.head(10)
```

```python
# d7 dataframe not considering the answers with no value
d7_df = pd.DataFrame()
d7_df['ML Projects'] = list(df['D7_Total_ML_Projects'])
```

```python
# remove those who didn't answered
d7_df = d7_df[d7_df['ML Projects'] != '-99']
```

```python
# convert column to integer - if you read before this line, you will get an string like '14'
d7_df = d7_df.astype({"ML Projects": int}, errors='raise') 
```

```python
single_boxplot(d7_df, "ML Projects",
               title="How many ML-enabled system projects have you participated in?")
```

```python
single_boxplot(d7_df, "ML Projects", outliers=False,
               title="How many ML-enabled system projects have you participated in?")
```

```python
single_violinplot(d7_df, "ML Projects",
                  title="How many ML-enabled system projects have you participated in?")
```

```python
single_histplot(d7_df, "ML Projects",
                title="How many ML-enabled system projects have you participated in?")
```

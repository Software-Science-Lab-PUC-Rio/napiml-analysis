```python
import pandas as pd
import numpy as np
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt
from ipynb.fs.full.utils import get_df, rename_values,remove_value
from ipynb.fs.full.plots import single_boxplot, single_histplot, single_violinplot
from ipynb.fs.full.bootstrapping_utils import bootstrapping_numerical_fields, confidence_interval
from statistics import mean, pstdev
```

```python
# get formatted and ready to use dataframe
df = get_df()
```

```python
# remove those who didn't answered
df = df[df['Q7_ML_Solution_Optimality_Extra_Effort'] != '-77']
```

```python
# convert column to integer - if you read before this line, you will get an string like '14'
df = df.astype({"Q7_ML_Solution_Optimality_Extra_Effort": int}, errors='raise') 
```

```python
valid_answers_list = list(df['Q7_ML_Solution_Optimality_Extra_Effort'])
```

```python
bootstrapped_list_data = bootstrapping_numerical_fields(answers=valid_answers_list, population_size=1000)
```

```python
lower, X_mean, upper = confidence_interval(data_points=bootstrapped_list_data)
print("Mean: " + str(X_mean) + "| Lower: " + str(lower) + "| Uper: " + str(upper))
```

```python
# d5 dataframe with only one column - plots purpose
q7_df = pd.DataFrame()
q7_df['ML Solution Optimality'] = bootstrapped_list_data
```

```python
single_boxplot(q7_df, "ML Solution Optimality",
               title="To what degree do you think that these “suboptimal” solutions cause added effort in the evolution of the ML-enabled system over time \n on a scale from 0% means no added effort/time needed, to 100% meaning a doubling of effort/time.")
```

```python
single_boxplot(q7_df, "ML Solution Optimality", outliers=False,
               title="To what degree do you think that these “suboptimal” solutions cause added effort in the evolution of the ML-enabled system over time \n on a scale from 0% means no added effort/time needed, to 100% meaning a doubling of effort/time.")
```

```python
#Which country do you currently work in?
import pandas as pd
from ipynb.fs.full.utils import get_df, rename_values,remove_value
from ipynb.fs.full.plots import single_barplot
```

```python
import matplotlib.pyplot as plt
import squarify # algorithm for treemap
```

```python
df = get_df()
df = df[['D2_Country']]
df.head(10)
```

```python
# remove undesirable value
remove_value(df,'0','D2_Country')
# index of our barplot
indexes = list(df['D2_Country'].value_counts().index)
# total answers of each index
values = list(df['D2_Country'].value_counts())
# dataframe specific for question D2
d2_df= pd.DataFrame(data={'Countries': indexes, 'Answers': values})
```

```python
single_barplot(d2_df,'Answers', 'Countries', 
               title="Which country do you currently work in?",
               x_label="Answers", y_label="Countries")
```

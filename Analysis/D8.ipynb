```python
import pandas as pd
from ipynb.fs.full.utils import get_df
from ipynb.fs.full.plots import single_boxplot, single_histplot, single_violinplot
```

```python
# get formatted and ready to use dataframe
df = get_df()
```

```python
df.head(10)
```

```python
# d8 dataframe not considering the answers with no value
d8_df = pd.DataFrame()
d8_df['ML Projects in Production'] = list(df['D8_ML_Production'])
```

```python
# let's have a quick overview about values in this column
d8_df['ML Projects in Production'].unique()
```

```python
# remove those who didn't answered
d8_df = d8_df[d8_df['ML Projects in Production'] != '-99']
```

```python
# convert column to integer - if you read before this line, you will get an string like '14'
d8_df = d8_df.astype({"ML Projects in Production": int}, errors='raise') 
```

```python
# compute total of answer considered in this question
total_answers = "N = " + str(len(d8_df['ML Projects in Production']))

single_boxplot(d8_df, "ML Projects in Production",
               title="Of all the ML-enabled system projects you have participated in,\n how many were actually deployed into a production environment?",
               total_answers=total_answers)
```

```python
# compute total of answer considered in this question
total_answers = "N = " + str(len(d8_df['ML Projects in Production']))

single_boxplot(d8_df, "ML Projects in Production", outliers=False,
               title="Of all the ML-enabled system projects you have participated in,\n how many were actually deployed into a production environment?",
               total_answers=total_answers)
```

```python
single_violinplot(d8_df, "ML Projects in Production",
                  title="Of all the ML-enabled system projects you have participated in, how many were actually deployed into a production environment?")
```

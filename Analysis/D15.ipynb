```python
algorithms = ['Apriori', 'Bayesian Networks', 'DBSCAN', 'Decision Trees', 'Ensembles', 'Gaussian Mixtures', 'K-Means', 'KNN', 
              'Lin/Poly Regression', 'Log Regression', 'Naive Bayes', 'Neural Networks', 'SVM']
```

```python
# the values follow the same position of previous declared list
algorithms_total = [
    df['D15_ML_Algorithms_Apriori'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Bayesian'].value_counts()['quoted'],
    df['D15_ML_Algorithms_DBSCAN'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Decision_Tree'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Ensembles'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Gaussian_Mixture'].value_counts()['quoted'],
    df['D15_ML_Algorithms_KMeans'].value_counts()['quoted'],
    df['D15_ML_Algorithms_KNN'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Linear_Regression'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Logistic_Regression'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Naive_Bayes'].value_counts()['quoted'],
    df['D15_ML_Algorithms_Neural_Networks'].value_counts()['quoted'],
    df['D15_ML_Algorithms_SVM'].value_counts()['quoted'],
]
```

```python
# dataframe specific for question D15
d15_df = pd.DataFrame(data={'Algorithms': algorithms, 'Answers': algorithms_total})
```

```python
d15_df
```

```python
# some people answered in the others field some of the listed options
extra_neural_networks = df['D15_ML_Algorithms_Others_Free'].value_counts()['GNN'] + df['D15_ML_Algorithms_Others_Free'].value_counts()['CNN'] + df['D15_ML_Algorithms_Others_Free'].value_counts()['Operational Neural Networks']
extra_regressions = df['D15_ML_Algorithms_Others_Free'].value_counts()['Ridge, LASSO/LARS Regressions']
```

```python
d15_df[d15_df['Algorithms'] == 'Neural Networks'] = ['Neural Networks', int(d15_df[d15_df['Algorithms'] == 'Neural Networks']['Answers']) + extra_neural_networks]
d15_df[d15_df['Algorithms'] == 'Lin/Poly Regression'] = ['Lin/Poly Regression', int(d15_df[d15_df['Algorithms'] == 'Lin/Poly Regression']['Answers']) + extra_regressions]
```

```python
# compute total of unique answers in this question (assuming the free answer field answers that we considered)
total = extra_neural_networks + extra_regressions
for idx, row in df.iterrows():
    if row['D15_ML_Algorithms_Apriori'] == 'quoted' or row['D15_ML_Algorithms_Bayesian'] == 'quoted' or row['D15_ML_Algorithms_DBSCAN'] == 'quoted' or row['D15_ML_Algorithms_Decision_Tree'] == 'quoted' or row['D15_ML_Algorithms_Ensembles'] == 'quoted' or row['D15_ML_Algorithms_Gaussian_Mixture'] == 'quoted' or row['D15_ML_Algorithms_KMeans'] == 'quoted' or row['D15_ML_Algorithms_KNN'] == 'quoted' or row['D15_ML_Algorithms_Linear_Regression'] == 'quoted' or row['D15_ML_Algorithms_Logistic_Regression'] == 'quoted' or row['D15_ML_Algorithms_Naive_Bayes'] == 'quoted' or row['D15_ML_Algorithms_Neural_Networks'] == 'quoted' or row['D15_ML_Algorithms_SVM'] == 'quoted':
        total += 1

total_answers = 'N = ' + str(total)
```

```python
# we sort the dataframe to plot the bars decreasingly
single_barplot(d15_df.sort_values(by=['Answers'], ascending=False), "Answers", "Algorithms", 
               title="What are the ML algorithms employed in the ML-enabled system projects in which you participated?",
               x_label="Answers", y_label="Algorithms", color="#429bf5",total_answers=total_answers)
```

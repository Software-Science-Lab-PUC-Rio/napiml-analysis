```python
import pandas as pd
from ipynb.fs.full.utils import custom_groupby, get_df, rename_values
from ipynb.fs.full.plots import multi_boxplot
```

```python
# get formatted and ready to use dataframe
df = get_df()
```

```python
df.head(10)
```

```python
# create a custom dataframe to handle only the columns that we want
d7_8_df = pd.DataFrame()
d7_8_df['ML Projects'] = list(df['D7_Total_ML_Projects'])
d7_8_df['ML Projects in Production'] = list(df['D8_ML_Production'])
```

```python
d7_8_df['ML Projects'].unique()
```

```python
d7_8_df['ML Projects in Production'].unique()
```

```python
# preprocessing d7 column
# remove those who didn't answered
d7_8_df = d7_8_df[d7_8_df['ML Projects'] != '-99']
```

```python
# preprocessing d8 column
# remove those who didn't answered
d7_8_df = d7_8_df[d7_8_df['ML Projects in Production'] != '-99']
```

```python
# convert column to integer - if you read before this line, you will get an string like '14'
d7_8_df = d7_8_df.astype({"ML Projects": int}, errors='raise') 
d7_8_df = d7_8_df.astype({"ML Projects in Production": int}, errors='raise') 
```

```python
def compute_differences(ml_projects, ml_projects_production):
    diffs = []
    for total, total_in_production in zip(ml_projects, ml_projects_production):
        # it doesn't make sense having a number of projects in production greater than the total projects of a participant
        if (total >= total_in_production):            
            if (total > 0):
                diffs.append((total_in_production / total) * 100)
            else:
                diffs.append(0)
        else:
            # in this case, we assume that once we have projects in production greater than
            # the number of projects a participant was involved, we set as 100%
            diffs.append(100)
    
    return diffs
```

```python
# add new column with the percentage of projects that a participant was involved and what went to production for real
d7_8_df['Percentage ML Projects in Production'] = compute_differences(d7_8_df['ML Projects'], d7_8_df['ML Projects in Production'])
```

```python
d7_8_df.head(10)
```

```python
d7_d8_box_df = pd.DataFrame()
d7_d8_box_df['Status'] = ['ML Project' for _ in range(len(list(d7_8_df['ML Projects'])))] + ['ML Projects in Production' for _ in range(len(list(d7_8_df['ML Projects in Production'])))] + ['Percentage ML Projects in Production' for _ in range(len(list(d7_8_df['Percentage ML Projects in Production'])))]
d7_d8_box_df['Total'] = [project for project in list(d7_8_df['ML Projects'])] + [prod for prod in list(d7_8_df['ML Projects in Production'])] + [perc for perc in list(d7_8_df['Percentage ML Projects in Production'])]
```

```python
d7_d8_box_df
```

```python
multi_boxplot(d7_d8_box_df, "Status", "Total", outliers=False, title='How many projects really went to production?')
```

```python
def rename_D3_values(d3_values):
    renamed_values = []
    rename_mapping = {
        '1-10 employees': '1-10',
        '11-50 employees': '11-50',
        '51-250 employees': '51-250',
        '251-500 employees': '251-500',
        '501-1,000 employees': '501-1000',
        '1,001-2,000 employees': '1001-2000',
        'More than 2,000 employees': '+2000'
    }
    
    for d3_value in d3_values:
        renamed_values.append(rename_mapping.get(d3_value, 'Not selected'))
    
    return renamed_values
```

```python
def generate_D3_barplot(df):
    indexes = list(df['D3_Company_Size'].value_counts().index)
    values = list(df['D3_Company_Size'].value_counts())
    d3_df= pd.DataFrame(data={
        'Company Size': rename_D3_values(indexes),
        'Answers': values})
    
    #draw a barplot with employees interval being x-axis and total of answers in that interval is y-axis
    d3_graphic = sns.catplot(data=d3_df, kind="bar", x="Company Size", y="Answers", ci="sd", palette="dark", alpha=.5, 
                             height=5, linewidth=5).set(title='What is the size of the organization you currently work for?')
    d3_graphic.set_xticklabels(rotation=45)
    d3_graphic.despine(left=True)
    d3_graphic.set_axis_labels("Total of Employees", "Answers")
```
